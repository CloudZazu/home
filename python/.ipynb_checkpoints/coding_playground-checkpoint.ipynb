```python
def find_all_paths_rec(root):
    if root is None:
        return []

    output = []

    if root.is_leaf():
        output.append(deque(root.value))
    else:
        output += find_all_paths_rec(root.left)
        output += find_all_paths_rec(root.right)

        for p in output:
            p.appendleft(root.value)

    return output
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
    
    def is_leaf(self):
        return self.left is None or self.right is None

def get_all_paths_topdown(root, all_paths, cur_path):
    
    if root is None:
        return
    
    cur_path.append(root.value)
    
    if root.is_leaf():
        all_paths.append(list(cur_path))
    else:
        get_all_paths_topdown(root.left, all_paths, cur_path)
        get_all_paths_topdown(root.right, all_paths, cur_path)
    
    cur_path.pop()
    
from collections import deque
# upwards convergence via deque
def find_all_paths_bottomup(root):
    all_paths = find_all_paths_rec(root)
    return [list(p) for p in all_paths]



def test(root):
    print(find_all_paths_bottomup(root))
    return

    all_paths = list()
    cur_path = list()
    get_all_paths_topdown(root, all_paths, cur_path)
    print(all_paths)

root = None
test(root)

root = TreeNode('a')
test(root)

root = TreeNode('b')
root.left = TreeNode('a')
root.right = TreeNode('c')
test(root)

root = TreeNode('d')
root.left = TreeNode('b')
root.left.left = TreeNode('a')
root.left.right = TreeNode('c')
root.right = TreeNode('f')
root.right.left = TreeNode('e')
root.right.right = TreeNode('g')
test(root)
```

```python
def find_startpoint(str1, str2):
    indexes = list()
    for idx in range(len(str2)):
        if str1[0] == str2[idx]:
            indexes.append(idx)
            
    return indexes   

def check_if_two_strs_are_equal_despite_shift(str1, str2):
    # abc, bca, cab
    # abcd, bcda, cdab
    
    starting_indexes = find_startpoint(str1, str2)
    
    if not starting_indexes:
        return False
    
    for starting_idx in starting_indexes:
        result = True
        
        for idx in range(len(str1)):
            start_point = starting_idx + idx

            if str1[idx] != str2[start_point % len(str2)]:
                result = False
                break
                
        if result:
            return True
            
    return False


def check_if_two_strs_are_equal_despite_shift2(str1, str2):
    
    if len(str1) != len(str2):
        return False
    
    tmp = str1 + str1
    
    idx2 = 0
    
    start = False
    for idx in range(len(tmp)):
        if tmp[idx] == str2[idx2]:
            start = True
        else:
            start = False
            
        if start:
            idx2 += 1
        else:
            idx2 = 0
            
        if idx2 == len(str2):
            return True

    return False
            
import time            
def test(str1, str2):
    start_time = time.time()
    result = check_if_two_strs_are_equal_despite_shift2(str1, str2)
    if not result:
        print(f'{str2} failed')
    elapsed_time = time.time() - start_time
    print(elapsed_time)

combinations = ['bcda','cdab','dabc','abcd','ffff']
for test_case in combinations:
    test('abcd', test_case)
    
test('aabc', 'caab') 
test('ddee','eedd')
test('ddff', 'ddee')
test('ddff', 'ddfe')
test('abcd', 'cdab')
```

```python
def riverSizes(matrix):
    # Write your code here.
    sizes = list()
    visited = [[False for value in row] for row in matrix]
    
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if visited[i][j]:
                continue
            if not matrix[i][j]:
                visited[i][j] = True
                continue
            
            river_size = bfs(i, j, matrix, visited)
            sizes.append(river_size)
                    
    return sizes
			
def bfs(x, y, matrix, visited):
    river_size = 1
    queue = list()
    queue.append((x,y))
    offsets = [(0, 1), (1, 0), (-1, 0), (0, -1)]
    while queue:
        cur_x, cur_y = queue.pop()
        visited[cur_x][cur_y] = True
        
        for x_offset, y_offset in offsets:
            valid = check_surroundings(cur_x+x_offset, cur_y+y_offset, matrix, visited)
            
            if not valid:
                continue
            
            river_size += 1
            visited[cur_x+x_offset][cur_y+y_offset] = True
            queue.append((cur_x+x_offset, cur_y+y_offset))
                    
    return river_size

def check_surroundings(x, y, matrix, visited):
    x_max = len(matrix)
    y_max = len(matrix[0])

    return 0 <= x < x_max and 0 <= y < y_max and not visited[x][y] and matrix[x][y]
```

```python
def numSubarraysWithSum(A, S):
    i_lo = i_hi = 0
    sum_lo = sum_hi = 0
    ans = 0
    for j, x in enumerate(A):
        # Maintain i_lo, sum_lo:
        # While the sum is too big, i_lo += 1
        sum_lo += x
        while i_lo < j and sum_lo > S:
            sum_lo -= A[i_lo]
            i_lo += 1

        # Maintain i_hi, sum_hi:
        # While the sum is too big, or equal and we can move, i_hi += 1
        sum_hi += x
        while i_hi < j and (
                sum_hi > S or sum_hi == S and not A[i_hi]):
            sum_hi -= A[i_hi]
            i_hi += 1
        
        if sum_lo == S:
            ans += i_hi - i_lo + 1
            
        print(f'i_lo: {i_lo} i_hi: {i_hi} sum_lo: {sum_lo} sum_hi: {sum_hi} ans: {ans}')
        
    return ans

import collections

def numSubarraysWithSum2(A, S):
    c = collections.Counter({0:1}) #memoization dictionary and key '0' is valued at '1' as we count the case when psum==S, we can also add this with res
    print(c)
    
    res, psum =0,0
    
    for v in A:
        psum += v # This basically checks the sum as we traverse the list
        res += c[psum-S] # This is keeping track of when we get past S, as when we get past S, we start with the new substring, and we already have the count and use it when we reach the SUM again and again
        c[psum] += 1
        print(f'v: {v} - c: {c} psum: {psum} psum-S: {psum-S} res: {res}')
    
    return res
    
A = [1,0,1,0,1]
S = 2

print(numSubarraysWithSum(A, S))
```

```python
# Moving average 

#Given a stream of integers and a window size, calculate the moving average of all integers in the sliding window.

class MovingAverage:
    def __init__(self, window_size=None):
        self.window_size = window_size
        self.buffer = list()
        
        self.currentSum = 0
        
    def next(self, val):
        if len(self.buffer) == self.window_size:
            self.currentSum -= self.buffer.pop(0)
            
        self.buffer.append(val)
        self.currentSum += val
         

        return self.currentSum / len(self.buffer)
    
mv_avg = MovingAverage(3)

arr = [1,2,3,4,5,4,3,2,1]

for val in arr:
    print(mv_avg.next(val))
```

```python
def find_string_anagrams(str, pattern):
  window_start, matched = 0, 0
  char_frequency = {}

  for chr in pattern:
    if chr not in char_frequency:
      char_frequency[chr] = 0
    char_frequency[chr] += 1

  result_indices = []
  # Our goal is to match all the characters from the 'char_frequency' with the current window
  # try to extend the range [window_start, window_end]
  for window_end in range(len(str)):
    right_char = str[window_end]
    if right_char in char_frequency:
      # Decrement the frequency of matched character
      char_frequency[right_char] -= 1
      if char_frequency[right_char] == 0:
        matched += 1

    if matched == len(char_frequency):  # Have we found an anagram?
      result_indices.append(window_start)

    # Shrink the sliding window
    if window_end >= len(pattern) - 1:
      left_char = str[window_start]
      window_start += 1
      if left_char in char_frequency:
        if char_frequency[left_char] == 0:
          matched -= 1  # Before putting the character back, decrement the matched count
        char_frequency[left_char] += 1  # Put the character back

  return result_indices


def main():
  print(find_string_anagrams("ppqp", "pq"))
  print(find_string_anagrams("abbcabc", "abc"))


main()
```

```python
def find_substring(str, pattern):
  
  lookup = dict()

  for char in pattern:
    if char not in lookup:
      lookup[char] = 0
    lookup[char] += 1

  window_start, match = 0, 0
  
  substr_start = 0
  min_size = float('inf')

  for window_end in range(len(str)):
    right_char = str[window_end]
    if right_char in lookup:
      lookup[right_char] -= 1
      if lookup[right_char] >= 0:
        match += 1

    while match == len(pattern): 
      current_size = window_end - window_start + 1
      if current_size < min_size:
        min_size = current_size
        substr_start = window_start 
      
      left_char = str[window_start]
      if left_char in lookup:
        
        if lookup[left_char] == 0:
          match -= 1
        lookup[left_char] += 1
      window_start += 1
      
  if min_size > len(str):
    return ''
  return str[substr_start:substr_start+min_size]

def main():
  print(find_substring("aabdec", "abc"))
  print(find_substring("abdabca", "abc"))
  print(find_substring("adcad", "abc"))

main()
```

```python
def find_substring(str, pattern):
  window_start, matched, substr_start = 0, 0, 0
  min_length = len(str) + 1
  char_frequency = {}

  for chr in pattern:
    if chr not in char_frequency:
      char_frequency[chr] = 0
    char_frequency[chr] += 1

  # try to extend the range [window_start, window_end]
  for window_end in range(len(str)):
    right_char = str[window_end]
    if right_char in char_frequency:
      char_frequency[right_char] -= 1
      if char_frequency[right_char] >= 0:  # Count every matching of a character
        matched += 1

    # Shrink the window if we can, finish as soon as we remove a matched character
    while matched == len(pattern):
      if min_length > window_end - window_start + 1:
        min_length = window_end - window_start + 1
        substr_start = window_start

      left_char = str[window_start]
      window_start += 1
      if left_char in char_frequency:
        # Note that we could have redundant matching characters, therefore we'll decrement the
        # matched count only when a useful occurrence of a matched character is going out of the window
        if char_frequency[left_char] == 0:
          matched -= 1
        char_frequency[left_char] += 1

  if min_length > len(str):
    return ""
  return str[substr_start:substr_start + min_length]


def main():
  print(find_substring("aabdec", "abc"))
  print(find_substring("abdabca", "abc"))
  print(find_substring("adcad", "abc"))
  print(find_substring("aabdecdabc",'abc'))

main()
```

```python
# Time - O(n); Space - O(1) since we assume the arr is sorted
def pair_with_targetsum(arr, target_sum):
    
    if len(arr) == 1:
        return arr[0] == target_sum
    
    if not len(arr):
        return [-1, -1]
    
    front, rear = 0, len(arr)-1
    
    while front < rear:
        total = arr[front] + arr[rear]
        
        if total < target_sum:
            front += 1
        elif total > target_sum:
            rear -= 1
        else:
            return [front, rear]
        
    return [-1, -1]

arr = [1,2,3,4,6]
print(pair_with_targetsum(arr, target_sum=5))
```

```python
# Problem 1: Given an unsorted array of numbers and a target ‘key’, remove all instances of ‘key’ in-place and return the new length of the array.
# Input: [3, 2, 3, 6, 3, 10, 9, 3], Key=3
# Output: 4

def removeKey(arr, key):
    nonDuplicate = 0
    
    i = 0
    while i < len(arr):
        if arr[i] != key:
            arr[nonDuplicate] = arr[i]
            nonDuplicate += 1
            
        i += 1

    return nonDuplicate

arr = [3, 2, 3, 6, 3, 10, 9, 3]
print(removeKey(arr, key=3))
```

```python
from collections import deque


def find_subarrays(arr, target):
  result = []
  product = 1
  left = 0
  for right in range(len(arr)):
    product *= arr[right]
    while (product >= target and left < len(arr)):
      product /= arr[left]
      left += 1
    # since the product of all numbers from left to right is less than the target therefore,
    # all subarrays from lef to right will have a product less than the target too; to avoid
    # duplicates, we will start with a subarray containing only arr[right] and then extend it
    temp_list = list()
    for i in range(right, left-1, -1):
      temp_list.append(arr[i])
      result.append(temp_list)
      print(f'Temp: {temp_list} Result: {result}')
  return result


def main():
  print(find_subarrays([2, 5, 3, 10], 30))
  print(find_subarrays([8, 2, 6, 5], 50))


main()
```

```python
# Determine the number of 1 bits in an integer
# Integer is about 2 billion of values
# 4 bytes 2^32 worht of values - 2^30 digits for positive and 2^30 for negative
def get_number_of_one_bits(integer):
    mask = 1
    
    one_bit_cnt = 0
    for i in range(32):
        if mask & integer:
            one_bit_cnt += 1
            
        mask <<= 1
        
    return one_bit_cnt

def get_number_of_one_bits(integer):
    
    one_bit_cnt = 0
    if integer == 0:
        return 0
    
    current_total = integer
    while current_total:
        current_total &= current_total-1
        one_bit_cnt += 1 
    
    return one_bit_cnt

get_number_of_one_bits(8)
```

```python
from typing import List

def heightChecker1(heights: List[int]) -> int:
    max_nr = max(heights)
    # initialize frequency array with 0's
    count = [0] * (max_nr + 1)
    # get frequencies
    for number in heights:
        count[number] += 1
    # create a sumcount array
    sumcount = [0] * (max_nr + 1)
    for index, number in enumerate(count[1:],start=1):
        sumcount[index] = number + sumcount[index-1]
        
    print(sumcount)
    
    # sumcount determines the index in sorted array
    # create output array
    output = [0] * len(heights)
    # loop backwards starting with last element for stable sort
    for p in range(len(heights)-1,-1,-1):
        output[sumcount[heights[p]]-1] = heights[p]
        sumcount[heights[p]] -= 1
        
    print(output)
    # return the difference compared to original array
    result = 0
    for index, number in enumerate(heights):
        if number != output[index]:
            result += 1
    return result

def heightChecker2(heights):
    return sum(h1 != h2 for h1, h2 in zip(heights, sorted(heights)))

heightChecker = heightChecker1

heights = [1,1,4,2,1,3]
heightChecker(heights)
```

```python
from typing import List

def move_zeroes(nums: List[int]):
    nonZeroIdx = 0
    for idx in range(len(nums)):
        if nums[idx]:
            nums[nonZeroIdx], nums[idx] = nums[idx], nums[nonZeroIdx]
            nonZeroIdx += 1

arr = [9,4,0,0,8,0,2]
move_zeroes(arr)
print(arr)
```

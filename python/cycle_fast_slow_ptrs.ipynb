```python
"""
Time complexity - O(n)
Space complexity - O(1)
Floyd's algorithm

Constraints: 
- array contains 1 to n 
- Only 1 duplicate - Can be repeated
- Array cant be modified
"""

def findDuplicateInt(arr):
    if not arr or len(arr) == 1:
        return -1
    
    fast = slow = arr[0]
    
    while slow != fast:
        slow = arr[slow]
        fast = arr[arr[fast]]

    slow = arr[0]
    
    while slow != fast:
        slow = arr[slow]
        fast = arr[fast]
        
    return slow
        
testArr = [3,1,2,4,3]
testArr = [1,1]
print(findDuplicateInt(testArr))
```
